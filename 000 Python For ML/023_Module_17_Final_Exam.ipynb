```python
# <--- STARTER CODE --->
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('final-employee-ds.csv')
```

### SOLUTION: Q-01

```python
training_hours = np.array(df['TrainingHours'])
training_hours = training_hours.astype(float)
```

```python
mean_training = training_hours.mean()
std_training = training_hours.std()

print(mean_training)
print(std_training)
```

### SOLUTION: Q-02

```python
df['Salary'].fillna(df['Salary'].mean(), inplace=True)
```

```python
req_employees = df.loc[(df['ExperienceYears'] > 12) & (df['ProjectCount'] > 5)]
```

```python
total_salary = req_employees['Salary'].sum()
print(total_salary)
```

### SOLUTION: Q-03

```python
filtered_employees = (((df['Department'] == 'IT') | (df['Department'] == 'Finance')) &
                      ((df['Age'] >= 30) & (df['Age'] <= 45)) &
                      (((df['PerformanceScore'] > 88))))

filtered_employees = df.loc[filtered_employees]

filtered_employees = filtered_employees.sort_values('Salary', ascending=False)

filtered_employees[['Name', 'Department', 'Salary', 'PerformanceScore']]
```

### SOLUTION: Q-04

```python
df['SalaryPerHour'] = df['Salary'] / (df['WorkHoursPerWeek'] * 4)

top_5 = df.sort_values('SalaryPerHour', ascending=False).head(5)

top_5[['Name', 'SalaryPerHour']]
```

### SOLUTION: Q-05

```python
dept_gen = df.groupby(['Department', 'Gender']).size()

pd.crosstab(df['Department'], df['Gender']).plot(kind='bar', color=['orange', 'red'], edgecolor='black')

plt.xticks(rotation=90)
plt.xlabel('Department')
plt.ylabel('Number of Employees')
plt.title('Employee Count By Department & Gender')
```
