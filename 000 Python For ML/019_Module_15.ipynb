# Module 15 â€” Data Visualization with Matplotlib

```python
# <--- Importing Required Libraries --->
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

## 2D Line Plots

```python
# <--- Basic Line Plot --->
hours = [2, 3, 4, 1, 5, 7, 3]
days = [1, 2, 3, 4, 5, 6, 7]

plt.plot(days, hours)
```

```python
# <--- Line Plot with Labels and Title --->
plt.xlabel('Days')
plt.ylabel('Study Hours')
plt.title('Trend of a student study time over a week')

plt.plot(days, hours)
```

```python
# <--- Multiple Lines in One Plot --->
hours_1 = [2, 3, 4, 1, 5, 7, 3]
hours_2 = [1, 4, 3, 5, 7, 2, 1]
days = [1, 2, 3, 4, 5, 6, 7]

plt.xlabel('Days')
plt.ylabel('Study Hours')
plt.title('Trend of student study time over a week')

plt.plot(days, hours_1)
plt.plot(days, hours_2)
```

## Line Plots from Pandas DataFrames

```python
# <--- Loading Data from CSV --->
df = pd.read_csv('enrollment_data.csv')
df
```

```python
# Plotting single column from DataFrame
plt.xlabel('Years')
plt.ylabel('Count of Enrollment')
plt.title('Student Enrollment over 8 years')

plt.plot(df['Year'], df['Programming'])
```

```python
# Plotting multiple columns from DataFrame
plt.xlabel('Years')
plt.ylabel('Count of Enrollment')
plt.title('Student Enrollment over 8 years')

plt.plot(df['Year'], df['Programming'])
plt.plot(df['Year'], df['Digital Marketing'])
```

```python
# <--- Adding Legend to Plot --->
plt.xlabel('Years')
plt.ylabel('Count of Enrollment')
plt.title('Student Enrollment over 8 years')

plt.plot(df['Year'], df['Programming'], label='Programming')
plt.plot(df['Year'], df['Digital Marketing'], label='Digital Marketing')

plt.legend(loc='best')  # Automatically finds the best position for legend
```

```python
# <--- Customizing Line Styles and Colors --->
plt.xlabel('Years')
plt.ylabel('Count of Enrollment')
plt.title('Student Enrollment over 8 years')

# Custom colors, line styles, markers, and sizes
plt.plot(df['Year'], df['Programming'], label='Programming',
         color='#11b86f', linewidth=3, linestyle='dashed',
         marker='o', markersize=7)
plt.plot(df['Year'], df['Digital Marketing'], label='Digital Marketing',
         color='black', linewidth=3, linestyle='dashed')

plt.legend()
```

```python
# <--- Adding Grid and Displaying Plot --->
plt.xlabel('Years')
plt.ylabel('Count of Enrollment')
plt.title('Student Enrollment over 8 years')

plt.plot(df['Year'], df['Programming'], label='Programming',
         color='#11b86f', linewidth=3, linestyle='dashed',
         marker='o', markersize=7)
plt.plot(df['Year'], df['Digital Marketing'], label='Digital Marketing',
         color='black', linewidth=3, linestyle='dashed')

plt.legend()
plt.grid()  # Add grid for better readability

plt.show()  # Display the plot
```

## Scatter Plots

```python
# <--- Loading Student IQ Data --->
students = pd.read_csv('student_IQdata.csv')
students
```

```python
# <--- Scatter Plot: Positive Correlation --->
# Relation between Study Hours and IQ Score - positive correlation

plt.xlabel('Study Hours')
plt.ylabel('IQ Score')
plt.title('Relation between Study Hours and IQ Score')

plt.scatter(students['Study_Hour'], students['IQ_Score'])
```

```python
# <--- Scatter Plot: No Correlation --->
# Relation between Shoe Size and IQ Score - no correlation

plt.xlabel('Shoe Size')
plt.ylabel('IQ Score')
plt.title('Relation between Shoe Size and IQ Score')

plt.scatter(students['Shoe_Size'], students['IQ_Score'])
```

```python
# <--- Scatter Plot: Negative Correlation --->
# Relation between Chilling Hours and IQ Score - negative correlation

plt.xlabel('Chilling Time')
plt.ylabel('IQ Score')
plt.title('Relation between Chilling Hours and IQ Score')

plt.scatter(students['Chilling_Hours'], students['IQ_Score'],
            marker='o', color='red')
```

## Histograms

```python
# <--- Histogram of IQ Scores --->
plt.hist(students['IQ_Score'], color='Red', edgecolor='black')
plt.xlabel('IQ Score')
plt.ylabel('Frequency of Students')
```

```python
# <--- Histogram with Custom Bins --->
# Using bins parameter to control number of bars
plt.hist(students['Study_Hour'], bins=15, color='Red', edgecolor='black')
plt.xlabel('Study Hours')
plt.ylabel('Frequency of Students')
```

```python
# <--- Histogram with Specific Bin Edges --->
# Using list to define specific bin boundaries
plt.hist(students['Study_Hour'], bins=[1, 2, 3, 4, 5, 6, 7, 8],
         color='Red', edgecolor='black')
plt.xlabel('Study Hours')
plt.ylabel('Frequency of Students')
```

## Bar Charts

```python
# <--- Creating Sample Data for Bar Charts --->
data = {
    'Student': ['B', 'C', 'D', 'E', 'F', 'G', 'H'],
    'Gender': ['Male', 'Female', 'Male', 'Female', 'Male', 'Female', 'Male'],
    'StudyHours': [2, 5, 3, 7, 4, 8, 1],
    'Attendance': [60, 88, 70, 95, 80, 98, 55],
    'Grade': ['Fail', 'Pass', 'Fail', 'Pass', 'Pass', 'Pass', 'Fail']
}

df = pd.DataFrame(data)
df
```

```python
# <--- Grouping Data by Gender --->
gender_group = df.groupby('Gender').size()

# Access group labels and counts
gender_group.index   # Output: Index(['Female', 'Male'], dtype='object')
gender_group.values  # Output: array([3, 4])
```

```python
# <--- Creating Bar Chart for Gender Distribution --->
plt.bar(gender_group.index, gender_group.values,
        color=['orange', 'red'], edgecolor='black')

plt.xlabel('Gender')
plt.ylabel('Count')
plt.title('Bar Chart of Gender among Students')
```

```python
# <--- Bar Chart on Real Data --->
df = pd.read_csv('student_data.csv')
df
```

```python
# <--- Bar Chart of Completion Status --->
status = df.groupby('CompletionStatus').size()

plt.bar(status.index, status.values,
        color=['green', 'orange', 'red'], edgecolor='black')

plt.xlabel('Completion Status')
plt.ylabel('Count')
plt.title('Bar Chart of Course Completion')
plt.show()
```

```python
# <--- Bar Chart of Student Locations --->
location = df.groupby('Location').size()

plt.bar(location.index, location.values, edgecolor='black')

plt.xlabel('Location')
plt.ylabel('Count')
plt.title('Bar Chart of Location of Students')
plt.show()
```

## Pie Charts

```python
# <--- Creating Pie Chart --->
status = df.groupby('CompletionStatus').size()

# autopct: display percentage, explode: separate a slice, shadow: add shadow effect
plt.pie(status, labels=status.index, autopct='%1.1f%%',
        explode=(0.1, 0, 0), shadow=True)
plt.show()
```
