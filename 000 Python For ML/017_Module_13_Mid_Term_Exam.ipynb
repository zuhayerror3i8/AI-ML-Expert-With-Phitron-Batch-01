```python
# <--- STARTER CODE --->
import numpy as np
import pandas as pd

np.random.seed(42)

ids = np.arange(1, 11)
ages = np.random.randint(18, 60, 10)
salaries = np.random.randint(30000, 90000, 10)
departments = np.array(["HR", "IT", "Finance", "IT", "HR", "Sales", "Finance", "IT", "Sales", "HR"])

DF = pd.DataFrame({
"id": ids,
"age": ages,
"salary": salaries,
"dept": departments
})

DF.to_csv("employees.csv", index=False)
print("Sample Data Created and Saved as employees.csv")
```

### SOLUTION: Q-01

```python
arr_age = np.array(ages)
arr_salary = np.array(salaries)

print(arr_age)
print(arr_age.dtype)
print(arr_age.ndim)
print(arr_age.shape)
print(arr_age.size)

print(arr_salary)
print(arr_salary.dtype)
print(arr_salary.ndim)
print(arr_salary.shape)
print(arr_salary.size)
```

### SOLUTION: Q-02

```python
highest_salary = np.max(arr_salary)
lowest_salary = np.min(arr_salary)
avg_salary = np.mean(arr_salary)
avg_age = np.mean(arr_age)

print(highest_salary)
print(lowest_salary)
print(avg_salary)
print(avg_age)
```

### SOLUTION: Q-03

```python
greater30 = arr_age[np.where(arr_age>30)]
greater30_cnt = np.count_nonzero(arr_age>30)

print(greater30)
print(greater30_cnt)
```

### SOLUTION: Q-04

```python
new_arr_age = arr_age + 5

print(new_arr_age)
```

### SOLUTION: Q-05

```python
total_salary = np.sum(arr_salary)

max_salary = np.max(arr_salary)
min_salary = np.min(arr_salary)

salary_diff = max_salary - min_salary

print(total_salary)
print(salary_diff)
```
