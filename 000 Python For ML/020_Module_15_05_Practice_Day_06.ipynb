```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('practice_day.csv')
df
```

## Problem 1: Add New Column and Save File

```python
df['Discounted_Price'] = df['Price'] * 0.9
df.to_csv('discounted_products.csv', index=False)
df
```

## Problem 2: Check Duplicates and Remove Them

```python
print("Duplicate count:", df['EmployeeID'].duplicated().sum())
df.drop_duplicates(subset='EmployeeID', inplace=True)
df
```

## Problem 3: Handle Missing Values

```python
print("Missing values:")
print(df[['Age', 'Salary']].isnull().sum())

df['Age'].fillna(df['Age'].mean(), inplace=True)
df.dropna(subset=['Salary'], inplace=True)
df
```

## Problem 4: Apply Statistical Functions

```python
print("Mean marks:")
print(df[['Marks1', 'Marks2', 'Marks3']].mean())

print("\nMax marks:")
print(df[['Marks1', 'Marks2', 'Marks3']].max())

df['Total_Marks'] = df['Marks1'] + df['Marks2'] + df['Marks3']
print("\nStandard deviation of total marks:", df['Total_Marks'].std())
df
```

## Problem 5: Histogram, Bar, Scatter, Pie

```python
plt.hist(df['Age'], bins=10, edgecolor='black')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.title('Histogram of Age')
plt.show()
```

```python
plt.scatter(df['Age'], df['Salary'])
plt.xlabel('Age')
plt.ylabel('Salary')
plt.title('Scatter Plot: Age vs Salary')
plt.show()
```

```python
plt.scatter(df['Price'], df['Discounted_Price'])
plt.xlabel('Price')
plt.ylabel('Discounted Price')
plt.title('Scatter Plot: Price vs Discounted Price')
plt.show()
```
